# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel("Customer Call List.xlsx")
df
```

**Here we delete all duplicate values**

```python
df = df.drop_duplicates()
df.tail()
```

**here we delete the column**

```python
df = df.drop(columns="Not_Useful_Column")
df
```

**Cleaning Last name**

```python
df["Last_Name"]= df["Last_Name"].str.strip("123./_")
df
```

**Cleaning Phone number**

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].str.replace('N/a', '', regex=False)
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if len(x) == 10 else x)
df
```

**Spliting Columns**

```python
df[["Street_Address","State","Pincode"]]=df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df.drop(columns="Address")
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
      df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
      df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == '':
      df.drop(x,inplace=True)
df
```

```python
df= df.reset_index(drop=True)
df
```
